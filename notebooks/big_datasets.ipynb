### Streaming datasets

```python
from datasets import load_dataset

dataset = load_dataset("openwebtext", streaming=True)
```

```python
next(iter(dataset['train']))
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
tokenized_dataset = dataset.map(lambda x: tokenizer(x['text']), batched=True)
```

```python
next(iter(tokenized_dataset['train']))
```

```python
### shuffling - shuffles elements in predefined buffer size (not entire dataset)
shuffled_dataset = dataset['train'].shuffle(buffer_size=10000, seed=42)
next(iter(shuffled_dataset))
```

```python
head = shuffled_dataset.take(5)
```

```python
list(head)
```

```python
### skipping first 10000 and getting a sample of 1000
train_dataset = shuffled_dataset.skip(10000)

validation_dataset = shuffled_dataset.take(1000)
```
